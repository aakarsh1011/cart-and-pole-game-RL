```python
#first reinforcement learning game
#till this point I know that it works on a rewrad and punishmment style 
#we have an ENVIRONMENT where we interface
#take STEPS which we want to do 
#take ACTIONS in the current situation of the game
#take OBSERVATION from the environment
#REWARD/PENALITY will be provided 
#check if the game is DONE OR NOT which gives an boolean
#INFO will be taken for debugging and see what the game is actually doing
```

```python
#importing packages we will need
import gym #provides environment for the game
import random #generates random values
import numpy as np #array calculation, if you don't know what it you are too early here
from tensorflow import keras #high level neural network API
```

```python
#intializing some variables 
env = gym.make('CartPole-v1')
env.reset()
goal_steps = 500
score_requirement = 60
initial_games = 10000
```

```python
#defing a function we will use to populate the data we need for our deep learning model training
def model_data_preparation():
    training_data = []
    accepted_scores = []
    for game_index in range(initial_games):#playing it 10000 times to collect sufficient data
        score = 0                     #initializing the current score
        game_memory =[]               #initializing the current game memory
        previous_observation = []     #initializing the previous observation of cart amd pole
        for step_index in range(goal_steps): #this loop is to play it 500 times which means succesfully completing the game
            action = random.randrange(0,2) #taking random actions to play the game 0--left, 1--right
            observation, reward, done, info = env.step(action) #after taking the action we will check the observation
                                                               #store it to previous observation and then check the score
                                                               #and reward and if the game is done or not
            if len(previous_observation) > 0: 
                game_memory.append([previous_observation, action])
                
            previous_observation = observation
            score += reward
            if done:
                break
            '''
            We will check whether this game fulfilling our minimum requirement or not 
            means are we able to play up to 60 steps or not
            If we are able to play 60 steps then we will add this score to accept_scores which we further print 
            to know how many games data and their score which we are feeding to our model
            '''
                
        if score >= score_requirement:
            accepted_scores.append(score)
            for data in game_memory:
                if data[1] == 1:            #encoding them if the value is 1--> moving right
                    output = [0, 1]
                elif data[1] == 0:      #if the value is 0 --> moving left
                    output = [1, 0]
                training_data.append([data[0], output])   #then saving it to our training data
                
                
        env.reset()     #resetting the env to make sure everything is clear to start again

    print(accepted_scores)   #this is to check how many games data and their score we are feeding to train
   
    return training_data        #finally return the triaing data
```

```python
training_data = model_data_preparation()
```

```python
'''def build_model(input_size, output_size):
    model = Sequential()
    model.add(Dense(128, input_dim=input_size, activation='relu'))
    model.add(Dense(52, activation='relu'))
    model.add(Dense(output_size, activation='linear'))
    model.compile(loss='mse', optimizer=Adam())

    return model'''
def build_model(input_size, output_size):
    model = keras.Sequential([
        keras.layers.Dense(128, input_dim=input_size, activation='relu'),           #128 is the number of nodes
        keras.layers.Dense(52, activation='relu'),
        keras.layers.Dense(output_size, activation='linear')
    ])
    model.compile(optimizer='adam', 
                  loss='mse',
                  metrics=['accuracy'])
    return model
```

```python
def train_model(training_data):
    X = np.array([i[0] for i in training_data]).reshape(-1, len(training_data[0][0]))
    y = np.array([i[1] for i in training_data]).reshape(-1, len(training_data[0][1]))
    model = build_model(input_size=len(X[0]), output_size=len(y[0]))
    
    model.fit(X, y, epochs=100)
    return model
```

```python
trained_model = train_model(training_data)
```

```python
scores = []
choices = []
for each_game in range(100):
    score = 0
    prev_obs = []
    for step_index in range(goal_steps):
        # Uncomment below line if you want to see how our bot is playing the game.
        env.render()
        if len(prev_obs)==0:
            action = random.randrange(0,2)
        else:
            action = np.argmax(trained_model.predict(prev_obs.reshape(-1, len(prev_obs)))[0])
        
        choices.append(action)
        new_observation, reward, done, info = env.step(action)
        prev_obs = new_observation
        score+=reward
        if done:
            break

    env.reset()
    scores.append(score)

print(scores)
print('Average Score:',sum(scores)/len(scores))
print('choice 1:{}  choice 0:{}'.format(choices.count(1)/len(choices),choices.count(0)/len(choices)))
```
